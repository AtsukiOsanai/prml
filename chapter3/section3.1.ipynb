### Polynomial Basis Function

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
M = 10
x = np.arange(-1, 1, 0.01)

def plot(x, vals):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    for val in vals:
        ax.plot(x, val)
    return plt
```

```python
def poly_basis(x, pow=1, bias=0):
    y = x ** pow
    return y
```

```python
poly_vals = [poly_basis(x, pow=m) for m in range(1, M)]
plt = plot(x, poly_vals)
plt.show()
```

### Gauss Basis Function

```python
def gauss_basis(x, mu, std):
    y = np.exp(-(x - mu) ** 2 / (2 * std ** 2))
    return y
```

```python
mus = np.arange(-1, 1, 0.2)
stds = 0.1 * np.ones(len(mus))
gauss_vals = [gauss_basis(x, mu, std) for mu, std in zip(mus, stds)]
```

```python
plt = plot(x, gauss_vals)
plt.show()
```

### Sigmoid Basis Function

```python
def sigmoid_basis(x, mu, scale):
    a = (x - mu) / scale
    y = 1 / (1 + np.exp(-a))
    return y
```

```python
mus = np.arange(-1, 1, 0.2)
scales = 0.1 * np.ones(len(mus))
sigmoid_vals = [sigmoid_basis(x, mu, scale) for mu, scale in zip(mus, scales)]
```

```python
plt = plot(x, sigmoid_vals)
plt.show()
```
